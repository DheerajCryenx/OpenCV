<h3>Exercise 3 - Coding Problem: Reading Videos</h3>
<p>In this exercise, we will be implementing all the concepts we have just learned, following the steps outlined in each block.</p>

```python
# Step-1 : Specify video capture object in the 'cap' variable with the path of any video you have.
path2 = "/home/dheerajt/Videos/test.mp4";
cap = cv2.VideoCapture(path2)

```

```python
# Step-2 : Initialize a while loop that runs as long as True is returned.
while(True):
    
    #Step-3 : Use the .read() command that returns ret and frame.
    
    ret, frame = cap.read()
    
    
    
    #Step-4 : Use cv2.imshow() within the loop to display each frame.
    
    
    cv2.imshow('image',frame);
   
    
    #Step-5 : Waitkey. cv2.waitKey(1) is used to display images in a frame of 1ms making it appear as a video
              #cv2.waitKey(0) is used to display images as along as a key is pressed.
    
    # waitkey also lets you control what FPS your video plays at
    
    # BONUS question: write code that plays any video at its native FPS
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):  #Allows you to quit the video pane when you press 'q'
        break
 
cap.release() 
cv2.destroyAllWindows() 
```
